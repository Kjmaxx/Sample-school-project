#### <font color=#8B0000>Problem B1: Special Relativity and Spacetime Diagrams in 2D [20 marks]</font>

In Special Relativity, the **Lorentz Transformation** is a linear transformation that describes how observations change when an observer is travelling at some velocity relative to their surroundings. 

In this problem we will consider an observer in 2D (one space dimension, $x$, and one time dimension, $t$). For simplicity, we will be using units such that $t$ has units of space, and our velocity $v$ will be in terms of the speed of light and is unitless (it's not important to the problem, but we do this by multiplying $t$ by the speed of light $c$ and defining $c=1$). Because our two directions are space and time, the tangents/slopes of lines represents a velocity. *Remember that a slope in a vector space corresponds to a specific vector direction.*

In the basis where a point in spacetime is represented by the vector $ ( t , x )^T$, a Lorentz Transformation for an observer at a velocity $v$ looks like: $$\Lambda = \begin{bmatrix} \gamma&-\gamma v \\ -\gamma v&\gamma \end{bmatrix}; \qquad \gamma=\dfrac{1}{\sqrt{1-v^2}}$$

If observer $A$ at rest sees some event at the point $ \vec{x}=\begin{bmatrix} t \\ x \end{bmatrix}$, a second observer $B$ in the same location travelling at velocity $v$ relative to them will observe that same event at the point $ \vec{x'}=\begin{bmatrix} t' \\ x' \end{bmatrix} = \Lambda \vec{x}$.
